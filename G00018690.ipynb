###  Introduction   

When it comes to algorithms, a sorting algorithm is a method for reorganizing a large number of items into a specific order, such as alphabetical, highest-to-lowest value or shortest-to-longest distance. These algorithms take lists of items as input data, perform specific operations on those lists and deliver ordered arrays as output. Sorting is essentially arranging a collection of items according to some pre-defined collection of rules.Sorting is often an important step as part of other computer algorithms, for example computer graphics are often layered on top of each other so a CG program may have t sort objects according to an above relation so that objects may be drawn from top to bottom. However it is also important in it's own right and not just as a pre-processing step for other taskd.  A simple example of sorting would be transactions in a bank account statement, sorted by transaction number or date. 


There are a number of conditions for sorting:

A collection of items is deemed to be sorted if each item in the collection is less than or equal to its successor. Duplicte elements must be contiguous in the resulting ordered collection and the sorted collectyion must be a permutation of the elements that originally formed it.


What sorting algoritim should be used?

Sorting algoritms just like any other algorithm have varying space and time efficiency. Time efficiency considers the time or number of operations required for the computer to run an algorithm. Space efficiency considers the amount of memory or storage the computer needs to run an algorithm. The efficiency of an algorithm can be measured by using either a priori analysis or a posteriori analysis. A priori analysis is where the efficiency is analysed by comparing the algorithms order of growth. Complexity is used to measure this by looking at factors such as the time needed to run the algorithm. A posteriori analysis looks at the efficiency of an algorithm that is implemented and run on a target platform. The efficiency of the algorithm is amnalysed by comparing the actual measurments collected during the experimentation and so performance is used to measure this type of analysis. Out of the two, complexity will impact performance but performance will not have an effect on complexity.

Typically complexity falls into one of a number of families as per below graph. If an algorithm is subdivided into two tasks, the poverall performance must be classified by the higher classification. It is also important to evaluate the class of input instance for which the algorithm exhibits its best, average and worst case runtimes.

#### Comparison Sort & Non-Comparison Sort

A comparison sort is a type of sorting algorithm which is a type of sorting algorithm which uses comparison operations only to determine which of the two elements should appear first in a sorted list. Examples of these algorithms include Bubble Sort, Insertion Sort and Selection Sort.

However sometimes algorithms with better worst-case times are needed and these are known as non-comparison sorting algorithms. Examples include Bucket Sort, Counting Sort and Radix Sort.

### 1. Bubble Sort 

Bubble Sort is a simple comparison based sorting algorithm which is used to sort a given set of n elements provided in the form of an array with n number of elements. It compares all of the elements in the array one by one and then sorts them based on their values. As it uses a constant amount of additiional working space in addition to the memory required for the input it is said to be an In-palce sorting algorithm and it is considered stable. Bubble Sort is simple to understand and implement but it is slow and impractical for most problems. This is why it is rearely used in real world problems and used more as an intoduction to the topic of sorting algorithms.

##### Properties:

In Bubble sort, n - 1 comparisons will be made in the first pass, n - 2 in the second, n - 3 in the third and so on hence the time complexity is O(n2) in the worst and average case performance. In the best case performance the time omplexity will be O(n) when the list is already sorted. The space complexity for Bubble sort is O(1) as only a single additional memory space is required.

-  Worst case performance O(n2)

-  Best case performance O(n)

-  Average case pergormance O(n2)

- Space complexity O(1)



##### Procedue:

1) Compare each element (except the last one) with it's neighbour to the right. If they are out of order, they are swapped which puts the largest element at the very end. The last element is now in the correct position.

2) Compare each element (except the last two) with its neighbour to the right. If they are out of order, they are swapped which puts the second last element next to the last. The last two elements are now in the correct position.

3) Compare each element (except the last three) with it's neighbour to the right.......This process is followed until the array is sorted.

### 2. Merge Sort

Merge Sort is an efficient recursive divide and conquer algorithm. It divides input array in two halves, calls itself for the two halves and then merges the two sorted halves. The algorithm is classed as stable and gives a good all-round performance. Merge sort is good for soting data with slow access times such as data that cannot be held in internal memory.


##### Properties:

Every time we divide an array into half in every step, it can be represented using the logarithmic function log n and the number of steps camnn be represented by log n + 1. To perform a single step operation to find out the middle of any subarray is O(1). To merge the subarrays made by dividing the original array a running time of O(n) is required. So the total time in Merge Sort will become n(log n + 1) which gives a time complexity of O(n*log n) in all cases. It requires an equal amount of additional space as the unsorted array so its not at all recommended for searching large unsorted arrays.

-  Worst case performance O(n log n)

-  Best case performance O(n log n)

-  Average case performance O(n log n)

-  Space complexity O(n)

##### Procedure:

1) If a list is larger than 0 or 1, seperate the list into two lists of equal or almost equal size and recursively sort the first and second halves seperately.

2) The two sorted halves are then mergred into one sorted list.

### 3. Counting Sort

Counting Sort is a stable non-comparison sorting technique based on keys between a specific range. It works by iterating through the input array, counting the number of times each item occurs, and using those counts to compute an items index in the final srted array. The algorithm assumes an input of size n where each item has a non-negative integer key with a reange of k. It is only suitable for direct se where the varuiation in keys is not significantly greater than the number of items. Counting sort is linear and assumes an integer k so it runs in O(n + k) in all cases.


##### Properties:

-  Worst case performance O(n + k)
-  Best case performance O(n + k)
-  Average case performance O(n + k)
-  Space complexity O(n + k) 


##### Procedures:

1)	Determine key range k of the input array.

2)	Initialize an array count of size k, which will be used to count the number of times that each key value appears in the input instance.

3)	Initialize an array result of size n, which will be used to store the sorted input.

4)	Iterate through the input array and record the number of times each distinct key value occurs in the input instance.

5)	Construct the sorted result array, based on the histogram of key frequencies stored in count. Refer to the ordering of leys in input to ensure that stability is preserved.

### 4. Insertion Sort

Insertion sort is a simple, in-pace, comparsion-based sorting algorithm. It works in a similar way to the method used by card players to sort cards in their hand. Here a sub-list is maintained which is always sorted. An element which is to be inserted in this sorted sub-list has to find its appropriate place and then must be inserted into the list. Insertion sort is easy to implement and works well on small lists and lists that are close to sorted. However it is very inefficient for large random lists. If an input array is alreadt sorted, the algorithm compare O(n) elements and performs no swaops and so it runs in O(n) for best case performance. The worst case peroformance will occur when the list is decreasing to order. As it's a stable sort the space complexity is O(1).

##### Properties:

-  Worst case performance O(n2)
-  Best case performance O(n)
-  Average case performance O(n2)
-  Space complexity O(1) 

##### Procedure:

1)	Start from the left of the array and set the “key” as the element at index 1. Move any elements to the left which are > the key right by one position and insert the “key”.

2)	Set the “key” as the element at index 2. Move any elements to the left which are > the key right by one position and insert the “key”.

3)	Set the “key” as the element at index 3. Move the elements to the left which are > the key right by one position and insert the “key”.

4)	Set the “key” as the element at index n-1. Move any elements to the left which are > the key right by one position and insert the key.

5)	Continue this process until no input elememts remain.

### 5. Selection Sort

Selection sort is a simple comparison-based in-place sorting algorithm. It sorts an array by repearedly finding the minimum or maximum element in an unsorted array and then putting it in its correct position in a sorted array. It is easy to implement but unstable. It gives a better performance than Bubble Sort but is still impractical for real world tasks with a significant input size so even in it's best case it runs in O(n2).

##### Properties:

-  Worst case performance O(n2)
-  Best case performance O(n2)
-  Average case performance O(n2)
-  Space complexity O(1) 

##### Procedure:

1)	Search elements 0 through n-1 and select the smallest. Swap it with the element in location 0

2)	Search elements 1 through n-1 and select the smallest. Swap it with the element in location 1

3)	Search elements 2 through n-1 and select the smallest. Swap it with the element in location 2

4)	Search elements 3 through n-1 and select the smallest. Swap it with the element in location 3

5)	Continue like this until there is nothing left to search.
